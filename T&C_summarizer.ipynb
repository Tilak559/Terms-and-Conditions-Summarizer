```python

class TextRank4Keyword():
    
    def __init__(self):
        self.d = 0.85 # damping coefficient, usually is .85
        self.min_diff = 1e-5 # convergence threshold
        self.steps = 10 # iteration steps
        self.node_weight = None # save keywords and its weight

    
    def set_stopwords(self, stopwords):  
        for word in STOP_WORDS.union(set(stopwords)):
            lexeme = nlp.vocab[word]
            lexeme.is_stop = True
    
    def sentence_segment(self, doc, candidate_pos, lower):
            sentences = []
            for sent in doc.sents:
                selected_words = []
                for token in sent:
                    if token.pos_ in candidate_pos and token.is_stop is False:
                        phrase = token.text
                        for adj in token.lefts:
                            if adj.pos_ == "ADJ":
                                phrase = adj.text + " " + phrase
                        for adj in token.rights:
                            if adj.pos_ == "ADJ":
                                phrase = phrase + " " + adj.text
                        if lower is True:
                            selected_words.append(phrase.lower())
                        else:
                            selected_words.append(phrase)
                sentences.append(selected_words)
            return sentences

        

    def get_vocab(self, sentences):
        vocab = OrderedDict()
        i = 0
        for sentence in sentences:
            for word in sentence:
                if word not in vocab:
                    vocab[word] = i
                    i += 1
            for word1, word2 in zip(sentence[:-1], sentence[1:]):
                bigram = f"{word1} {word2}"
                if bigram not in vocab:
                    vocab[bigram] = i
                    i += 1
        return vocab

    
    def get_token_pairs(self, window_size, sentences, phrase_length = 1):
        token_pairs = list()
        for sentence in sentences:
            for i, word in enumerate(sentence):
                for j in range(i+1, i+window_size):
                    if j >= len(sentence):
                        break
                    pair = (word, sentence[j])
                    if pair not in token_pairs:
                        token_pairs.append(pair)
                    if phrase_length > 1 and j-i == phrase_length:
                        phrase = ' '.join(sentence[i:j+1])
                        if phrase not in token_pairs:
                            token_pairs.append(phrase)
        return token_pairs
        
    def symmetrize(self, a):
        return a + a.T - np.diag(a.diagonal())
    
    def get_matrix(self, vocab, token_pairs):
        vocab_size = len(vocab)
        g = np.zeros((vocab_size, vocab_size), dtype='float')
        for word1, word2 in token_pairs:
            i, j = vocab[word1], vocab[word2]
            g[i][j] = 1
            
        g = self.symmetrize(g)
        
        norm = np.sum(g, axis=0)
        g_norm = np.divide(g, norm, where=norm!=0)
        
        return g_norm

    
    def get_keywords(self, number=10):
        node_weight = OrderedDict(sorted(self.node_weight.items(), key=lambda t: t[1], reverse=True))
        for i, (key, value) in enumerate(node_weight.items()):
            print(key + ' - ' + str(value))
            if i > number:
                break
        
        
    def analyze(self, text, 
                candidate_pos=['NOUN', 'PROPN'], 
                window_size=4, lower=False, stopwords=list()):
        
        self.set_stopwords(stopwords)
        
        doc = nlp(text)
        
        sentences = self.sentence_segment(doc, candidate_pos, lower) # list of list of words
        
        vocab = self.get_vocab(sentences)
        
        token_pairs = self.get_token_pairs(window_size, sentences)
        
        g = self.get_matrix(vocab, token_pairs)
        
        pr = np.array([1] * len(vocab))
        
        previous_pr = 0
        for epoch in range(self.steps):
            pr = (1-self.d) + self.d * np.dot(g, pr)
            if abs(previous_pr - sum(pr))  < self.min_diff:
                break
            else:
                previous_pr = sum(pr)

        node_weight = dict()
        for word, index in vocab.items():
            node_weight[word] = pr[index]
        self.node_weight = node_weight
```

```python
import re
from collections import Counter
import nltk

class TextAnalyzer:
#     def __init__(self):
#         nltk.download('punkt')

    def extract_keywords(self, text):
        sentence_list = re.split(r'[.!?]+', text)

        word_list = [re.findall(r'\w+', sentence.lower()) for sentence in sentence_list]

        word_freq = Counter()
        for sentence in word_list:
            for word in sentence:
                word_freq[word] += 1

        word_scores = Counter()
        for sentence in word_list:
            for word in sentence:
                word_scores[word] += word_freq[word]

        phrase_scores = Counter()
        for sentence in sentence_list:
            phrase_list = re.findall(r'\b[\w\s]+\b', sentence.lower())
            for phrase in phrase_list:
                phrase_scores[phrase.strip()] += sum(word_scores[word] for word in re.findall(r'\w+', phrase))

        keywords = [phrase for phrase, score in phrase_scores.most_common(20)]

        return keywords

    def form_paragraph(self, text):
        sentences = nltk.sent_tokenize(text)

        paragraph = sentences[0].capitalize()

        proper_nouns = []

        for sentence in sentences[1:]:
            words = nltk.word_tokenize(sentence)

            tagged_words = nltk.pos_tag(words)

            proper_nouns += [word for word, pos in tagged_words if pos == 'NNP']

            if sentence.split()[0] in proper_nouns:
                paragraph += "\n\n" + sentence.capitalize()
            elif sentences[sentences.index(sentence)-1][-1] in ['.', '!', '?'] or sentences[sentences.index(sentence)-1][-2:] in ['."','!"','?"']:
                paragraph += " " + sentence.capitalize()
            else:
                paragraph += " " + sentence

        return paragraph
```
